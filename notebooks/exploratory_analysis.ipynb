```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '../data/processed/kanye_lyrics_with_features.csv'
data = pd.read_csv(file_path)

data['lyrics_length'] = data['lyrics'].apply(len)

#distribution of lyrics lengths
plt.figure(figsize=(10, 6))
plt.hist(data['lyrics_length'], bins=50, color='skyblue', edgecolor='black')
plt.title('Lyrics Length Distribution')
plt.xlabel('Lyrics Length')
plt.ylabel('Frequency')
plt.show()
```

```python
data['lyric_length'] = data['lyrics'].apply(len)

#plot
plt.figure(figsize=(10, 6))
plt.hist(data['lyric_length'], bins=30, color='skyblue', edgecolor='black')
plt.title('Lyric Length Distribution')
plt.xlabel('Lyric Length (Characters)')
plt.ylabel('Number of Songs')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
tokenized_lyrics = data['lyrics'].apply(word_tokenize)
```

```python
type_token_ratios = tokenized_lyrics.apply(lambda x: len(set(x)) / len(x) if len(x) > 0 else 0)
```

```python
#diversity Index
simpsons_diversity_index = tokenized_lyrics.apply(lambda x: 1 - sum((freq / len(x))**2 for freq in FreqDist(x).values()))
```

```python
data['type_token_ratio'] = type_token_ratios
data['simpsons_diversity_index'] = simpsons_diversity_index
```
